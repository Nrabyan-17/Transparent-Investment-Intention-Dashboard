## 7. Validasi Integritas Antar Tabel

Cek apakah ID yang ada di satu tabel juga ada di tabel lain (untuk memastikan join nantinya tidak menghasilkan banyak null).

```python
# customerID di transactions vs di customers
customer_ids_trans = set(transactions['customerID'].unique())
customer_ids_master = set(customers['customerID'].unique())

print(f"Unique customerID di customers: {len(customer_ids_master):,}")
print(f"Unique customerID di transactions: {len(customer_ids_trans):,}")
print(f"customerID di transactions yang TIDAK ada di customers: {len(customer_ids_trans - customer_ids_master):,}")
print(f"customerID di customers yang TIDAK ada di transactions: {len(customer_ids_master - customer_ids_trans):,}")
```

```python
# ISIN di transactions vs di assets
isin_trans = set(transactions['ISIN'].unique())
isin_assets = set(assets['ISIN'].unique())

print(f"Unique ISIN di assets: {len(isin_assets):,}")
print(f"Unique ISIN di transactions: {len(isin_trans):,}")
print(f"ISIN di transactions yang TIDAK ada di assets: {len(isin_trans - isin_assets):,}")
```
